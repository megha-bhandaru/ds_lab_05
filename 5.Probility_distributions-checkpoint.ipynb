# Normal Distribution

### 1.	Generate the 10000 random numbers using stanadrd  normal distribution mean =0 and variance =1, and plot the normal distributions using seaborn.



```python
import seaborn as sns
from scipy.stats import norm
import numpy as np
np.random.seed(32)
data= norm.rvs(size=10000,loc=0,scale=1)
print(data[0:11])

```

```python
ax = sns.distplot(data,
                  bins=100,
                  kde=True,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Normal Distribution', ylabel='Frequency')
```

# Bernoulli Distributions

### 2.	Generate 10000, bernoulli random numbers with success probability p =0.3,  and plot the bernoulli distributions using seaborn.

```python
# import bernoulli
from scipy.stats import bernoulli

# generate bernoulli
data= bernoulli.rvs(size=10000,p=0.3)
print(data[0:11])
ax= sns.distplot(data,
                 kde=False,
                 color="skyblue",
                 hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Bernoulli', ylabel='Frequency')

```

# Binomial Distributions

### 3.	Generate 10000, Binomial   random numbers with success probability p =0.5 and number of trail n=10 and plot the Binomial  distributions using seaborn.

```python
from scipy.stats import binom
data = binom.rvs(n=10,p=0.5,size=10000)
ax = sns.distplot(data,
                  kde=False,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Binomial', ylabel='Frequency')
```

# Poisson distribution

### 4.	Generate 10000 random numbers from poisson random variable with mu = 0.3 and plot them.

```python
from scipy.stats import poisson
data_poisson = poisson.rvs(mu=3, size=10000)
ax = sns.distplot(data_poisson,
                  kde=False,
                  color='green',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Poisson', ylabel='Frequency')

```
