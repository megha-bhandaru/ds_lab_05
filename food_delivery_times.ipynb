```python
from google.colab import files
uploaded = files.upload()

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('Food_Delivery_Times.csv')

print("--- Data Head ---")
print(df.head())
print("\n--- Data Info ---")
df.info()

df['Courier_Experience_yrs'].fillna(df['Courier_Experience_yrs'].mean(), inplace=True)
df['Weather'].fillna(df['Weather'].mode()[0], inplace=True)
df['Traffic_Level'].fillna(df['Traffic_Level'].mode()[0], inplace=True)
df['Time_of_Day'].fillna(df['Time_of_Day'].mode()[0], inplace=True)

sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Distance_km', y='Delivery_Time_min', alpha=0.6)
plt.title('Delivery Time Increases with Distance', fontsize=16)
plt.xlabel('Distance (km)', fontsize=12)
plt.ylabel('Delivery Time (minutes)', fontsize=12)
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Traffic_Level', y='Delivery_Time_min', order=['Low', 'Medium', 'High'])
plt.title('Impact of Traffic on Delivery Time', fontsize=16)
plt.xlabel('Traffic Level', fontsize=12)
plt.ylabel('Delivery Time (minutes)', fontsize=12)
plt.show()

plt.figure(figsize=(12, 7))
sns.barplot(data=df, x='Weather', y='Delivery_Time_min', ci=None, palette='mako')
plt.title('Impact of Weather on Delivery Time', fontsize=16)
plt.xlabel('Weather Condition', fontsize=12)
plt.ylabel('Average Delivery Time (minutes)', fontsize=12)
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='Vehicle_Type', y='Delivery_Time_min', palette='viridis')
plt.title('Delivery Time by Vehicle Type', fontsize=16)
plt.xlabel('Vehicle Type', fontsize=12)
plt.ylabel('Delivery Time (minutes)', fontsize=12)
plt.show()

plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='Courier_Experience_yrs', y='Delivery_Time_min', scatter_kws={'alpha': 0.4}, line_kws={'color': 'red'})
plt.title('Experienced Couriers are Faster', fontsize=16)
plt.xlabel('Courier Experience (years)', fontsize=12)
plt.ylabel('Delivery Time (minutes)', fontsize=12)
plt.show()

plt.figure(figsize=(10, 8))
numerical_cols = ['Distance_km', 'Preparation_Time_min', 'Courier_Experience_yrs', 'Delivery_Time_min']
correlation_matrix = df[numerical_cols].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Between Numerical Features', fontsize=16)
plt.show()

print("\nAnalysis complete.")
```
